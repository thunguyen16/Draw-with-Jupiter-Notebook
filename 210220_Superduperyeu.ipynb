```python
import turtle as t
def rectangle(hor,ver,col):
    t.pendown() #tạo con trỏ
    t.pensize(1)
    t.color(col)
    t.begin_fill()
    for counter in range(1,3):
        t.forward(hor)
        t.right(90)
        t.forward(ver)
        t.right(90)
    t.end_fill()
    t.penup()
t.penup()
t.speed('slow')
t.bgcolor('Dodger blue')

#Viết chữ E
t.goto(-500,150)
rectangle(20,150,'pink')
t.goto(-500,150)
rectangle(60,30,'pink')
t.goto(-500,90)
rectangle(60,30,'pink')
t.goto(-500,30)
rectangle(60,30,'pink')

#Viết chữ M
t.goto(-420,150)
rectangle(20,150,'pink')
t.goto(-400,150)
def parallel1(hei,bas,col):
    t.pendown() #tạo con trỏ
    t.pensize(1)
    t.color(col)
    t.begin_fill()
    for counter in range(1,3):
        t.forward(hei)
        t.right(45)
        t.forward(bas)
        t.right(135)
    t.end_fill()
    t.penup()
t.goto(-420,150)
parallel1(20,60,'pink')
def parallel2(hei,bas,col):
    t.pendown() #tạo con trỏ
    t.pensize(1)
    t.color(col)
    t.begin_fill()
    for counter in range(1,3):
        t.forward(hei)
        t.right(135)
        t.forward(bas)
        t.right(45)
    t.end_fill()
    t.penup()
t.goto(-330,150)
parallel2(20,60,'pink')
t.goto(-330,150)
rectangle(20,150,'pink')

#Viết chữ Y
t.goto(-270,150)
parallel1(30,60,'pink')
t.goto(-180,150)
parallel2(30,60,'pink')
t.goto(-225,108)
rectangle(30,108,'pink')

#Viết chữ Ê:
t.goto(-140,150)
rectangle(20,150,'pink')
t.goto(-140,150)
rectangle(60,30,'pink')
t.goto(-140,90)
rectangle(60,30,'pink')
t.goto(-140,30)
rectangle(60,30,'pink')
t.goto(-120,175)
parallel2(15,25,'pink')
t.goto(-115,175)
parallel1(15,25,'pink')

#Viết chữ U:
t.goto(-60,150)
rectangle(20,120,'pink')
t.goto(-60,30)
parallel1(20,35,'pink')
t.goto(-30,15)
rectangle(20,10,'pink')
t.goto(0,30)
parallel2(20,35,'pink')
t.goto(0,150)
rectangle(20,120,'pink')

#Viết chữ A:
def hbh1(ngh,ng,col):
    t.pendown() #tạo con trỏ
    t.pensize(1)
    t.color(col)
    t.begin_fill()
    for counter in range(1,3):
        t.forward(ngh)
        t.right(80)
        t.forward(ng)
        t.right(100)
    t.end_fill()
    t.penup()
def hbh2(ngh,ng,col):
    t.pendown() #tạo con trỏ
    t.pensize(1)
    t.color(col)
    t.begin_fill()
    for counter in range(1,3):
        t.forward(ngh)
        t.right(100)
        t.forward(ng)
        t.right(80)
    t.end_fill()
    t.penup()
t.goto(100,150)
hbh2(20,150,'pink')
t.goto(100,150)
rectangle(30,10,'pink')
t.goto(110,150)
hbh1(20,150,'pink')
t.goto(100,75)
rectangle(37,20,'pink')

#Viết chữ N:
t.goto(170,150)
rectangle(20,150,'pink')
def hbh3(hei,bas,col):
    t.pendown() #tạo con trỏ
    t.pensize(1)
    t.color(col)
    t.begin_fill()
    for counter in range(1,3):
        t.forward(hei)
        t.right(60)
        t.forward(bas)
        t.right(120)
    t.end_fill()
    t.penup()
t.goto(170,150)
hbh3(20,172,'pink')
t.goto(255,150)
rectangle(20,150,'pink')

#Viết chữ H:
t.goto(300,150)
rectangle(20,150,'pink')
t.goto(300,90)
rectangle(80,20,'pink')
t.goto(380,150)
rectangle(20,150,'pink')
t.hideturtle()

```
